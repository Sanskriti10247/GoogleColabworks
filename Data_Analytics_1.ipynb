```python
ratings=pd.read_csv('/content/ratings.csv')
ratings.head()
```

```python
ratings[['UserID','MovieID']].duplicated().sum()
```

```python
ratings=ratings.sort_values(by='Timestamp')
ratings.head()
```

```python
ratings.drop_duplicates(subset=['UserID','MovieID'],keep='last',inplace=True)
ratings.info()
```

```python
ratings.groupby('MovieID')['Rating'].mean().reset_index()
```

```python
ratings_agg= ratings.groupby('MovieID').agg({'Rating':'mean','Timestamp':'count'}).reset_index()
```

```python
ratings_agg.rename(columns={'Rating':'avg_rating','Timestamp':'rating_count'},inplace=True)

ratings_agg.head()
```

```python
pd.merge(movies1,ratings,on='MovieID',how='left')
```
